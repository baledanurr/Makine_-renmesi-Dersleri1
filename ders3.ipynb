```python
X=digits.data
y=digits.target
print(X.shape)
print(y.shape)
```

```python
from sklearn.manifold import Isomap
```

```python
iso=Isomap(n_neighbors=2,n_components=2)
```

```python
iso.fit(X)
```

```python
data2=iso.transform(X)
data2.shape
```

```python
plt.scatter(data2[:,0],data2[:,1],c=digits.target,edgecolor='none',alpha=0.5,cmap=plt.cm.get_cmap('tab10',10))
plt.colorbar(label='digit label',ticks=range(10))
```

```python
from sklearn.model_selection import train_test_split
X_egitim,X_test,y_egitim,y_test=train_test_split(X,y,random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model=GaussianNB()
```

```python
model.fit(X_egitim,y_egitim)
```

```python
y_model=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_model)
```

```python
from sklearn.metrics import confusion_matrix
matris=confusion_matrix(y_test,y_model)
```

```python
sns.heatmap(matris,square=True,annot=True,cbar=False)
plt.xlabel('tahmin değer')
plt.ylabel('gerçek değer')
```
